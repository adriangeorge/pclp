## Secțiune recapitulativă din laboratoarele precedente

În această secțiune, vom recapitula conceptele învățate în laboratoarele anterioare: comentarii, variabile, operații aritmetice, comparații, operatori binari și funcții din pachetul `math`.

### Exemple recapitulative

```python

# Definirea unor variabile
numar_intreg = 10
numar_float = 3.14
text = "Salut"

# Operatii aritmetice simple
suma = numar_intreg + 5
diferenta = numar_float - 1.5
produs = numar_intreg * numar_float
impartire = numar_intreg / 2

print(f"Suma: {suma}")
print(f"Diferenta: {diferenta}")
print(f"Produsul: {produs}")
print(f"Impartirea: {impartire}")
```

```python
# Comparatii
x = 15
y = 10

print(f"x > y: {x > y}")
print(f"x < y: {x < y}")
print(f"x == y: {x == y}")
print(f"x != y: {x != y}")

# Operatori binari (exemple simple)
a = 5  # 0101 în binar
b = 3  # 0011 în binar

print(f"a & b (AND): {a & b}")
print(f"a | b (OR): {a | b}")
print(f"a ^ b (XOR): {a ^ b}")
print(f"~a (NOT): {~a}") # Complementul pe biți (atenție la reprezentarea numerelor negative)
print(f"a << 1 (Shift stânga): {a << 1}") # Înmulțire cu 2
print(f"a >> 1 (Shift dreapta): {a >> 1}") # Împărțire la 2 (diviziune întreagă)
```

```python
# Utilizarea pachetului math
import math

raza = 5
arie_cerc = math.pi * raza**2
radical = math.sqrt(25)
sinus = math.sin(math.pi/2)

print(f"Aria cercului cu raza {raza}: {arie_cerc}")
print(f"Radical din 25: {radical}")
print(f"Sinus de pi/2: {sinus}")
```

 Putem efectua diverse operații pe ele.

### Exemple operații pe șiruri de caractere

### Funcții utile:
- `len(str)` – returnează lungimea unui string
- `str.upper()` – convertește toate literele la majuscule
- `str.lower()` – convertește toate literele la minuscule
- `str.title()` – transformă prima literă din fiecare cuvânt în majusculă
- `str.find(substring)` – caută poziția primei apariții a unui substring
- `str.replace(old, new)` – înlocuiește un substring cu altul


```python
text = "Universitatea tehnică din București"
print(len(text))          # Lungimea textului
print(text.upper())       # Toate majuscule
print(text.lower())       # Toate minuscule
print(text.title())       # Prima literă majusculă în fiecare cuvânt
print(text.find('Tehnică'))  # Poziția unde se găsește 'Tehnică'
print(text.replace('București', ''))

# print(text)
textModificat = text.title()
print(text)
print(textModificat)
```

```python
# Definirea unui string
salut = "Bună ziua!"

# Concatenarea (unirea) string-urilor
nume = "Andrei"
mesaj = salut + " " + nume
print(mesaj)

# Repetarea string-urilor
repetat = "Ha" * 3
print(repetat)

# Lungimea unui string
lungime_mesaj = len(mesaj)
print(f"Lungimea mesajului: {lungime_mesaj}")
print("Lungimea mesajului:", lungime_mesaj)
```

## Indexare și „slicing” în string-uri

Caracterele dintr-un string sunt indexate începând de la `0`. Poți accesa un caracter sau o porțiune dintr-un string folosind **indexarea** și **tăierea (slicing)**.


```python
text = 'Python'
print(text[0])   # Primul caracter
print(text[-1])  # Ultimul caracter
print(text[1:4]) # Substring (caracterele de la indexul 1 la 3)
print(text[:3])  # Primele trei caractere
print(text[3:])  # De la al patrulea caracter până la final
```

```python
# Accesarea caracterelor individuale (indexare)
primul_caracter = salut[0]
al_treilea_caracter = salut[2]
print(f"Primul caracter: {primul_caracter}")
print(f"Al treilea caracter: {al_treilea_caracter}")

# Slicing (extragerea sub-șirurilor)
sub_sir = mesaj[0:5] # Caracterele de la indexul 0 la 4 (exclusiv 5)
print(f"Sub-șir: {sub_sir}")

sub_sir_pana_la_sfarsit = mesaj[6:] # De la indexul 6 până la sfârșit
print(f"Sub-șir până la sfârșit: {sub_sir_pana_la_sfarsit}")

sub_sir_de_la_inceput = mesaj[:5] # De la început până la indexul 4 (exclusiv 5)
print(f"Sub-șir de la început: {sub_sir_de_la_inceput}")

# Slicing cu pas (step)
sub_sir_pas = mesaj[::2] # Fiecare al doilea caracter
print(f"Sub-șir cu pas: {sub_sir_pas}")
```

### Funcții utile pentru șiruri de caractere

```python
text = "Python este un limbaj de programare"

# Convertire la majuscule/minuscule
majuscule = text.upper()
minuscule = text.lower()
print(f"Majuscule: {majuscule}")
print(f"Minuscule: {minuscule}")

# Verificarea dacă un string începe/se termină cu un anumit sub-șir
incepe = text.startswith("Python")
se_termina = text.endswith("programare")
print(f"Începe cu 'Python': {incepe}")
print(f"Se termină cu 'programare': {se_termina}")

# Găsirea indexului unui sub-șir
index_limbaj = text.find("limbaj")
print(f"Indexul 'limbaj': {index_limbaj}")

# Înlocuirea unui sub-șir
inlocuit = text.replace("limbaj de programare", "simplu și puternic")
print(f"Text înlocuit: {inlocuit}")

# Împărțirea unui string (split)
cuvinte = text.split(" ")
print(f"Cuvinte (rezultat split): {cuvinte}")

# Verificarea tipului de conținut (is...())
numere_str = "12345"
litere_str = "abcde"

print(f"'12345' este numeric? {numere_str.isnumeric()}")
print(f"'abcde' este alfabetic? {litere_str.isalpha()}")
```
