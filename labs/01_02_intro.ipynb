### 3. Variabilele

În Python, variabile se pot declara unde este nevoie de ele. O variabilă este creată în momentul în care i se asignează o valoare.

Python permite declararea variabilelor dinamic, fără tip explicit. Tipuri uzuale: `int`, `float`, `str`, `bool`.

Variabilele nu au tip de date la momentul declarării:
- Variabilele iau implicit tipul de date al valorii
- Variabilele pot să schimbe tipul de date
- Dacă vrem să forțăm un tip de date pentru o variabilă, putem să facem asta cu operatorul cast

```python
# Exemplu de creare și folosire a variabilelor în Python

# Variabile inițiale
nume = "Ana"
varsta = 20
media_notelor = 9.5
is_student = True

# Afișarea valorii și tipului fiecărei variabile
print("Nume:", nume, "- Tip:", type(nume))
print("Vârstă:", varsta, "- Tip:", type(varsta))
print("Media notelor:", media_notelor, "- Tip:", type(media_notelor))
print("Este student?", is_student, "- Tip:", type(is_student))

```

### 6. Operatori binari

Operatorii binari sunt operatori care acționează asupra a doi operanzi. În Python, aceștia sunt foarte utili pentru diferite operații, fie aritmetice, fie bitwise.

Operări bitwise (acționează asupra biților)
- AND (&)
- OR (|)
- XOR (^)
- Shiftare stânga (<<)
- Shiftare dreapta (>>)

```python
a = 10  # În binar: 1010
b = 4   # În binar: 0100

# Bitwise AND: returnează 1 dacă ambii biți sunt 1, altfel 0
print("a & b =", a & b)  # 1010 & 0100 = 0000 (0)

# Bitwise OR: returnează 1 dacă cel puțin un bit este 1
print("a | b =", a | b)  # 1010 | 0100 = 1110 (14)

# Bitwise XOR: returnează 1 dacă biții sunt diferiți
print("a ^ b =", a ^ b)  # 1010 ^ 0100 = 1110 (14)

# Bitwise NOT: inversează toți biții (complement)
print("~a =", ~a)  # ~00001010 = 11110101 (-11 în complement față de 2)

# Shiftare la stânga (multiplicare cu 2 la puterea shift-ului)
print("a << 1 =", a << 1)  # 10100 (20)

# Shiftare la dreapta (împărțire la 2 la puterea shift-ului)
print("a >> 2 =", a >> 2)  # 0010 (2)
```

### 7. Funcții

Python grupează colecții de module (fișiere cu cod Sursă) care conțin funcții în pachete (biblioteci).

Exemple de funcții:
- Pachetul math oferă funcții matematice bazate pe bibliotecile de tip C și este optimizat pentru viteză.
- Funcția input([message]) (Parametrul message este opțional și ne ajută să afișă um mesaj)



```python
import math

# Constante matematice
print("Pi:", math.pi)           # Constanta π
print("e:", math.e)             # Constanta e

# Rădăcina pătrată
print("Rădăcina pătrată a 16:", math.sqrt(16))

# Exponențiale și logaritmi
print("exp(2):", math.exp(2))   # e la puterea 2
print("log(100, 10):", math.log(100, 10)) # logaritm pe bază 10

# Factorial
print("5! =", math.factorial(5))

# Funcții trigonometrice
print("sin(90°):", math.sin(math.radians(90)))
print("cos(0):", math.cos(0))

# Rotunjiri speciale
print("floor(3.7):", math.floor(3.7))   # Partea întreagă inferioară
print("ceil(3.2):", math.ceil(3.2))     # Partea întreagă superioară

# Modulul unui număr
print("fabs(-10.5):", math.fabs(-10.5))

# Putere
print("2 la puterea 8:", math.pow(2, 8))
```
